```python
def normalise(number):
  # how many places have we shifted?
  shift = 0
  # strip leading bits
  pos = number.find(".")
  while pos > 1 and number[0] == number[1]:
    number = number[0] + number[2:] + "0"
    pos = number.find(".")
  # shift point up to the second position
  while pos > 1:
    number = number[:pos-1] + "." + number[pos-1] + number[pos+1:]
    shift += 1
    pos = number.find(".")
  while number[0] == number[2]:
    number = number[:2] + number[3:] + "0"
    shift -= 1
  return (number, shift)

def binary(denary):
  integer = int(denary)
  fraction = denary - integer
  binary = "."
  column = 0
  bits = m_length + 3
  # convert integer part (left of the point) using bitwise logic
  while integer >= 2**column:
    binary = str(int((integer & 2**column)/2**column)) + binary
    column += 1
  binary = "0" + binary
  # convert the fraction part using place value
  if fraction == 0:
    binary += "0"
  else:
    column = 0.5
    while abs(fraction) > 0.0000000000001:
      if fraction >= column:
        binary += "1"
        fraction -= column
      else:
        binary += "0"
      column /= 2
  if len(binary) < bits:
    binary = "0"*(bits - len(binary)) + binary
  return binary

# number of bits used to represent the mantissa and exponent
e_length = 5
m_length = 16
# ask for the denary value
denary = eval(input("Denary value: "))
if (denary < 0) :
  sign = '.1.'
else:
  sign = '.0.'
if denary != 0:
  # convert to binary
  bin_val = binary(abs(denary))
  print("\nThe binary representation of", abs(denary), "is", bin_val)
  # normalise the mantissa
  normalised = normalise(bin_val)
  print("This can be normalised to", normalised[0][:m_length+1], "by shifting", abs(normalised[1]), "place(s) to the", ("right.","left.")[normalised[1]<1])
  # convert the exponent to binary
  exponent = binary(normalised[1]+2**(e_length-1))[:-2]
  exponent = ("0"*(e_length-len(exponent)) + exponent)[e_length*-1:]
  print("The exponent is therefore", normalised[1])
  print("\nUsing", m_length , "bits for the mantissa and", e_length,"for the exponent (these can be changed in the code):")
  # trim the mantissa to the required number of bits
  print("Machine representation:",exponent+sign+normalised[0][2:m_length+1])
  # check whether the value has been truncated
  if abs(normalised[1]) >= 2**(e_length-1):
    print("This number requires an exponent that needs more than",e_length,"bits.")
else:
  print("That's just zero, so the mantissa and exponent are both zero.")
```
